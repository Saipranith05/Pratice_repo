```python
import pandas as pd
```

```python
df = pd.read_csv('C:/Users/Akshitha/New folder/pranith_class/tips.csv')
```

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
x = df.loc[:,["total_bill"]]
```

```python
y = df.loc[:,["tip"]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
# we need to predict by using x_train
y_pred = lr.predict(x_test)
```

```python
# we need to compare the y_pred and the y_test

from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(y_test,y_pred)
```

```python
import numpy as np
```

```python
rmse = np.sqrt(mse)
```

```python
import gradio as gr

# Assuming lr is your trained Linear Regression model
def tip_prediction(total_bill):
    print(f"Received total_bill: {total_bill}")  # Debugging line
    
    if not isinstance(total_bill, (int, float)):
        return "Please provide a valid number."
    
    try:
        predicted_tip = int(lr.predict([[total_bill]]))
        print(f"Predicted tip: {predicted_tip}")  # Debugging line
        return predicted_tip
    except Exception as e:
        return f"Error: {str(e)}"  # Return the error as a string

# Create the Gradio interface
pranith = gr.Interface(fn=tip_prediction, inputs="number", outputs="number", title="Tip Prediction")

# Launch the interface
pranith.launch()
```
